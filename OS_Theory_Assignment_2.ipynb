**Optimal Page Replacement Algorithm**

```python
#sample string: 70120304230321201701

def findTheNoneIndex(frame):
  for i in range(0,len(frame)):
    if(frame[i] == None):
      return i

def checkIfPageIsNoLongerExists(page,frame,curr_position):
  for i in range(0,len(frame)):
    if frame[i] not in page[curr_position : len(page)]:
      return i
  return -1

def checkIfFrameExists(frame,page):
  if(page in frame):
    return True
  else:
    return False

def checkIfNone(frame):
  flag = False
  for i in range(0,len(frame)):
    if(frame[i] == None):
      flag = True
  return flag

def findNotUsedPage(page,frame,curr_position):
  value = checkIfPageIsNoLongerExists(page,frame,curr_position)
  if(value != -1):
    return value
  else:
    position = [None] * len(frame)
    for i in range(curr_position,len(page)):
      for j in range(0 ,len(frame)):
        if(frame[j] == page[i]):
          position[j] = i
      if(checkIfNone(position) == False):
        break
    x = position.index(max(position))
    return x



string = input("Enter the string")
frame_size = input("Enter frame size")

page = []
frame = [None] * int(frame_size)

for i in range(0, len(string)):
  page.append(int(string[i]))

# page = [7,0,1,2,0,3,0,4,2,3,0,3,2,1,2,0,1,7,0,1]
# frame = [None] * 4
hit = 0
miss = 0

for i in range(0,len(page)):
  # print("")
  # print("FOR:  ",page[i])
  # print("-----------------------")
  if(checkIfNone(frame) == True):
    if(checkIfFrameExists(frame, page[i]) == True):
      print(frame)
      hit += 1
      continue
    frame[findTheNoneIndex(frame)] = page[i]
    miss += 1
    print(frame)
  else:

    # print("Inside else")
    if(checkIfFrameExists(frame, page[i]) == True):
      print(frame)
      hit += 1
      continue
    else:
      notUsedPage = findNotUsedPage(page,frame,i)
      frame[notUsedPage] = page[i]
      miss += 1
      print(frame)

print("MISS/Page fault:", miss)
print("HIT:", hit)

```

**Least Recently Used (LPU) Page Replacement Algorithm**

```python
def checkIfNone(frame):
  flag = False
  for i in range(0,len(frame)):
    if(frame[i] == None):
      flag = True
  return flag


def checkIfFrameExists(frame,page):
  if(page in frame):
    return True
  else:
    return False


def findTheNoneIndex(frame):
  for i in range(0,len(frame)):
    if(frame[i] == None):
      return i



def findNotUsedPage(page,frame,curr_position):
  position = [None] * len(frame)
  for i in range(curr_position,-1,-1):
    for j in range(0 ,len(frame)):
      if(frame[j] == page[i]):
        position[j] = i
    if(checkIfNone(position) == False):
      break

  # print(position)
  x = position.index(min(position))
  return x


string = input("Enter the string")
frame_size = input("Enter frame size")

page = []
frame = [None] * int(frame_size)

for i in range(0, len(string)):
  page.append(int(string[i]))

# page = [7,0,1,2,0,3,0,4,2,3,0,3,2,1,2,0,1,7,0,1]
# frame = [None] * 3
hit = 0
miss = 0

for i in range(0,len(page)):
  if(checkIfNone(frame) == True):
    if(checkIfFrameExists(frame, page[i]) == True):
      print(frame)
      hit += 1
      continue
    frame[findTheNoneIndex(frame)] = page[i]
    miss += 1
    print(frame)
  else:
    # print("Inside else")
    if(checkIfFrameExists(frame, page[i]) == True):
      print(frame)
      hit += 1
      continue
    else:
      notUsedPage = findNotUsedPage(page,frame,i)
      frame[notUsedPage] = page[i]
      miss += 1
      print(frame)

print("MISS/Page fault:", miss)
print("HIT:", hit)

```

**First In First Out Page Replacement Algorithm**

```python
def checkIfNone(frame):
  flag = False
  for i in range(0,len(frame)):
    if(frame[i] == None):
      flag = True
  return flag

def checkIfFrameExists(frame,page):
  if(page in frame):
    return True
  else:
    return False


def findTheNoneIndex(frame):
  for i in range(0,len(frame)):
    if(frame[i] == None):
      return i

def findFrameIndex(frame,value):
  for i in range(0, len(frame)):
    if(frame[i] == value):
      return i 


string = input("Enter the string")
frame_size = input("Enter frame size")

page = []
frame = [None] * int(frame_size)

for i in range(0, len(string)):
  page.append(int(string[i]))

# page = [7,0,1,2,0,3,0,4,2,3,0,3,2,1,2,0,1,7,0,1]
# frame = [None] * 3
queue = []
hit = 0
miss = 0

for i in range(0,len(page)):

  if(checkIfNone(frame) == True):
    if(checkIfFrameExists(frame, page[i]) == True):
      print(frame)
      hit += 1
      continue
    frame[findTheNoneIndex(frame)] = page[i]
    queue.append(page[i])
    miss += 1
    print(frame)

  else:
    if(checkIfFrameExists(frame, page[i]) == True):
      print(frame)
      hit += 1
      continue
    else:
      first_come = queue.pop(0)
      index = findFrameIndex(frame,first_come)
      frame[index] = page[i]
      queue.append(page[i])
      miss +=1
      print(frame)


print("MISS/Page fault: ", miss)
print("HIT: ", hit)
```
